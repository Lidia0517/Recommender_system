Demographic Filtering
--
Եթե որպես գնահատական (score) վերցնենք ուղղակի ֆիլմի միջին վարկանիշը (average rating), ապա արդյունքը կլինի անարդար և ոչ օբյեկտիվ։ Հետևաբար որպես լուծում վերցրել եմ IMDb-ի կշռված վարկանիշը (weighted rating)՝
photo_5291867624344067077_x.jpg

- $v$ — ֆիլմի ստացած ձայների (votes) քանակն է
- $m$ — ձայների այն նվազագույն քանակն է, որը պահանջվում է ցուցակում ընդգրկվելու համար (listed in the chart)
- $R$ — տվյալ ֆիլմի միջին գնահատականն է (average rating)
- $C$ — ամբողջ բազայի ֆիլմերի միջին գնահատականն է(mean vote across the whole report)
Քանի որ մեր տվյալներում կա $v$ (vote_count) և $R$ (vote_average), հաշվենք $C$

```python
C = movies.vote_average.mean()
C
```

```python
m = movies['vote_count'].quantile(0.9)
m
```

Սահմանենք, որ ֆիլմը պետք է գտնվի 90 պերցենտիլում, ինչը նշանակում է, որ դիտարկվելու համար այն պետք է ավելի շատ ձայն ունենա, քան ցուցակի ֆիլմերի առնվազն 90%-ը (ֆիլտրում է շատ քիչ ձայներ ունեցող ֆիլմերը, որոնք կարող էին ունենալ մոլորեցնող բարձր կամ ցածր միջին գնահատականներ)


```python
qualified_movies = movies.copy().loc[movies.vote_count >= m]

print(qualified_movies.shape)
qualified_movies.sample(5)
#Կա 481 ֆիլմ, որոնք համապատասխանում են այս ցուցակում ընդգրկվելու պահանջներին
```

```python
def weighted_rating(df_row, m=m, C=C):
    v = df_row.vote_count
    R = df_row.vote_average

    #Հաշվարկը ըստ բանաձևի
    return (v / (v + m) * R) + (m / (m + v) * C)
```

```python
#Սահմանենք նոր սյունակ՝'score' և հաշվարկենք դրա արժեքը weighted_rating() ֆունկցիայի միջոցով
qualified_movies['score'] = qualified_movies.apply(weighted_rating, axis='columns')
qualified_movies.score[:10]
```

```python
#Դասավորում ըստ score-ի
qualified_movies = qualified_movies.sort_values('score', ascending=False)
```

```python
qualified_movies[['title', 'vote_average', 'score']].head(15)
```

```python
top_10_movies = qualified_movies.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='score', y='title', data=top_10_movies, palette='viridis')
plt.xlim(7.5, 8.5)
plt.xlabel('Weighted Score')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies via Demographic Filtering')
plt.show()
```

Օգտագործենք NLTK գրադարանի PorterStemmer-ը՝ բառերը իրենց արմատական ձևին բերելու համար (օրինակ՝ "loving", "loved", "loves" → "love")։ stems() ֆունկցիան կիրառում է սա tags տեքստի (string) յուրաքանչյուր բառի վրա։ Սա օգնում է մոդելին՝ նույն բառի տարբեր գրվածքով դիտարկել որպես իրար հավասար

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stems(text):
  l = []
  for i in text.split():
    l.append(ps.stem(i))

  return " ".join(l)
```

# Predicting Movie Revenue with KNN

Այս հատվածում, անցնում ենք *recommendation* դեպի *prediction*: Կարո՞ղ ենք կանխատեսել ֆիլմի **եկամուտը**՝ օգտագործելով դրա մյուս բնութագրերը(budget, runtime, popularity, vote count, vote average, genres, release year, etc.)

Խնդրին մոտենալու եք հետևյալ ճանապարհով.

**Classification** — կանխատեսում է արդյոք ֆիլմը **"High Revenue"** **Medium Revenue** **"Low Revenue"**. Գնահատվում է օգտագործելով **Accuracy, Precision, Recall**, և **Confusion Matrix**.



## 1. Տվյալների պատրաստում (Անկախ և կախյալ փոփոխականներ)

**Կախյալ տվյալները (target, y):** `revenue`

**Անկախ տվյալները (features, x):** թվային հատկանիշներ, որոնք հավանական է, որ ազդում են ֆիլմի եկամտի վրա:
- `budget`
- `popularity`
- `runtime`
- `vote_average`
- `vote_count`
- `release_year` (վերցված `release_date`)
- `genre_count` (ժանրերի քանակը)
- `cast_size`

Առաջին հերթին մեզ անհրաժեշտ է մաքրել մի քանի բան՝ որոշ ֆիլմեր ունեն budget = 0 կամ runtime = 0, որոնք ամենայն հավանականությամբ բացակայող արժեքներ են (missing values), այլ ոչ թե իրական զրոներ, և մեզ անհրաժեշտ է նորից վերլուծել (re-parse) genres-ը (քանի որ movies-ն այստեղ սկզբնական dataframe-ն է, այլ ոչ թե new_df-ը).

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,confusion_matrix, classification_report)
import numpy as np
import pandas as pd
import ast

credits['cast_size'] = credits['cast'].apply(lambda s: len(s) if pd.notna(s) else 0)

df = movies.merge(credits[['movie_id', 'cast_size']], left_on='id', right_on='movie_id', how='left')

def parse_genres(text):
    try:
        return [g['name'] for g in ast.literal_eval(text)]
    except (ValueError, SyntaxError):
        return []

df['genres_list'] = df['genres'].apply(parse_genres)
df['genre_count'] = df['genres_list'].str.len()
top_genres = ['Drama', 'Comedy', 'Thriller', 'Action', 'Adventure', 'Romance',
              'Crime', 'Science Fiction', 'Family', 'Fantasy']

for g in top_genres:
    col = 'genre_' + g.replace(' ', '_')
    df[col] = df['genres_list'].apply(lambda lst: int(g in lst))

df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year

#Զրոների փոխարինում
clean_cols = ['budget', 'runtime', 'revenue']
df[clean_cols] = df[clean_cols].replace(0, np.nan)

# Սյուների սահմանում և մաքրում
feature_cols = ['budget', 'popularity', 'runtime', 'vote_average',
                'vote_count', 'release_year', 'genre_count', 'cast_size']
target_col = 'revenue'

df = df.dropna(subset=feature_cols + [target_col])

print(f"Rows remaining after cleaning: {df.shape[0]}")
print(df[['title', 'budget', 'revenue', 'release_year', 'genre_count', 'cast_size']]
      .head().to_string(index=False))
```

Correlation heatmap մեզ օգնում է տեսնել, թե որ թվային հատկանիշներն են ամենաշատը կապված revenue-ի (եկամտի) հետ՝ նախքան մեր մոդելները կառուցելը

```python
numeric_cols = ['revenue', 'budget', 'runtime', 'popularity', 'vote_count',
                'vote_average', 'release_year', 'genre_count', 'cast_size']
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation matrix of numeric features")
plt.tight_layout()
plt.show()

print(corr['revenue'].sort_values(ascending=False))
```

  - Եկամուտը կապակցված է բյուջեի, ձայների քանակի և հանրաճանաչության հետ ամենաշատը, իսկ օրինակ թողարկման տարվա կամ ժանրերի քանակի հետ այդքան էլ ոչ
  - Նաև կարելի նկատել որ ձայների քանակը և հանրաճանաչությունն են լավ կապակցված

## Դասակարգում (High Revenue, Medium Revenue, Low Revenue)


```python
df['revenue_class'] = pd.qcut(df['revenue'], q=3, labels=['Low', 'Medium', 'High'])
print(df['revenue_class'].value_counts())

X_knn = df[feature_cols]
y_knn = df['revenue_class']

X_train_k, X_test_k, y_train_k, y_test_k = train_test_split(
    X_knn, y_knn, test_size=0.2, random_state=42, stratify=y_knn)

scaler = StandardScaler()
X_train_k_scaled = scaler.fit_transform(X_train_k)
X_test_k_scaled = scaler.transform(X_test_k)

print("\nTraining set:", X_train_k.shape)
print("Test set    :", X_test_k.shape)
```

Ընտրեք k-ն՝ չափելով accuracy-ն տարբեր K-երի դեպքում

```python
k_values = range(1, 26)
accuracies = []
for k in k_values:
    knn_k = KNeighborsClassifier(n_neighbors=k)
    knn_k.fit(X_train_k_scaled, y_train_k)
    accuracies.append(accuracy_score(y_test_k, knn_k.predict(X_test_k_scaled)))

plt.figure(figsize=(8, 5))
plt.plot(list(k_values), accuracies, marker='o')
plt.xlabel('k (number of neighbors)')
plt.ylabel('Test accuracy')
plt.title('KNN Accuracy vs. k')
plt.tight_layout()
plt.show()

best_k = list(k_values)[int(np.argmax(accuracies))]
print(f"Best k: {best_k}  (accuracy = {max(accuracies):.3f})")
```

Մարզեք KNN մոդելը և կանխատեսումներ կատարեք թեստային բազմության վրա

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_k_scaled, y_train_k)
y_pred_k = knn.predict(X_test_k_scaled)

preview_k = pd.DataFrame({'Actual': y_test_k.values[:10], 'Predicted': y_pred_k[:10]})
print(preview_k.to_string(index=False))
```

### Գնահատումը: Accuracy, Precision, Recall

- **Accuracy**: Ընդհանուր ճշգրտություն
- **Precision**: Բոլոր այն ֆիլմերից, որոնց մոդելը կանխատեսել էր որպես «Բարձր եկամտաբեր», որի մասն էր իրականում այդպիսին
- **Recall**: Բոլոր այն ֆիլմերից, որոնք իրականում «Բարձր եկամտաբեր» էին, որի մասին մոդելը կարողացավ ճիշտ հայտնաբերել
- **F1-score** - Precision-ի  և Recall-ի հավասարակշռված միջինն է

```python
acc = accuracy_score(y_test_k, y_pred_k)
prec = precision_score(y_test_k, y_pred_k, average='macro')
rec = recall_score(y_test_k, y_pred_k, average='macro')

print(f"Accuracy          : {acc:.3f}")
print(f"Precision : {prec:.3f}")
print(f"Recall   : {rec:.3f}\n")
print("Per-class report:")
print(classification_report(y_test_k, y_pred_k))
```

Ընդհանուր ճշգրտություն - մոդելը ճիշտ է գուշակում դեպքերի 66.6%-ում

High (0.75) և Low (0.73) - Մոդելը բավականին լավ է տարբերակում «ծայրահեղ» արժեքները (բարձր և ցածր եկամուտ ունեցողներին)

Medium (0.50): Մոդելը շատ վատ է աշխատում միջին եկամտաբերություն ունեցողներին գտնելիս։ Դրա precision և recall ցուցանիշները շատ ցածր են համեմատած մյուսների հետ:

### Confusion Matrix

```python
labels = ['Low', 'Medium', 'High']
cm = confusion_matrix(y_test_k, y_pred_k, labels=labels)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title(f'Confusion Matrix (k={best_k})')
plt.tight_layout()
plt.show()

print(cm)
```
